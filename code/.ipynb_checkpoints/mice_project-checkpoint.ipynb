## Уровень каких белков различается в мышиной модели синдрома Дауна
В 2015 году на мышиной модели были проведены исследования как синдром Дауна влияет на изменения уровней различных белков. Данные для проекта можно скачать по следующей ссылке -- https://archive.ics.uci.edu/ml/datasets/Mice+Protein+Expression#  
Doi статьи на которую можно ориентироваться -- 10.1371/journal.pone.0119491.  

### Dataset information
The data set consists of the expression levels of 77 proteins/protein modifications that produced detectable signals in the nuclear fraction of cortex. There are 38 control mice and 34 trisomic mice (Down syndrome), for a total of 72 mice. In the experiments, 15 measurements were registered of each protein per sample/mouse. Therefore, for control mice, there are 38x15, or 570 measurements, and for trisomic mice, there are 34x15, or 510 measurements. The dataset contains a total of 1080 measurements per protein. Each measurement can be considered as an independent sample/mouse.   
  
The eight classes of mice are described based on features such as genotype, behavior and treatment. According to genotype, mice can be control or trisomic. According to behavior, some mice have been stimulated to learn (context-shock) and others have not (shock-context) and in order to assess the effect of the drug memantine in recovering the ability to learn in trisomic mice, some mice have been injected with the drug and others have not.   
  
Classes:   
c-CS-s: control mice, stimulated to learn, injected with saline (9 mice)   
c-CS-m: control mice, stimulated to learn, injected with memantine (10 mice)   
c-SC-s: control mice, not stimulated to learn, injected with saline (9 mice)   
c-SC-m: control mice, not stimulated to learn, injected with memantine (10 mice)   
  
t-CS-s: trisomy mice, stimulated to learn, injected with saline (7 mice)   
t-CS-m: trisomy mice, stimulated to learn, injected with memantine (9 mice)   
t-SC-s: trisomy mice, not stimulated to learn, injected with saline (9 mice)   
t-SC-m: trisomy mice, not stimulated to learn, injected with memantine (9 mice)   
  
The aim is to identify subsets of proteins that are discriminant between the classes.

### 3. Построим линейную модель, способную предсказать уровень продукции белка ERBB4_N на основании данных о других белках в эксперименте 

```python
# data for analysis
X = mice_data_cleaned.iloc[:, 1:-4].drop(columns=['ERBB4_N'])
y = mice_data_cleaned['ERBB4_N']
```

```python
# сделаем а ля one-hot encoding для Genotype, Treatment, Behavior чтобы были поправки на разные группы
X['trisomy']    = mice_data_cleaned['Genotype'] == 'Ts65Dn'
X['memantine']  = mice_data_cleaned['Treatment'] == 'Memantine'
X['stimulated'] = mice_data_cleaned['Behavior'] == 'C/S'

X['trisomy']    = X['trisomy'] .astype(int)
X['memantine']  = X['memantine'].astype(int)
X['stimulated'] = X['stimulated'].astype(int)
```

```python
# data split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
print(len(y_train), len(y_test))
```

```python
# создание и обучение модели
reg_model = LinearRegression().fit(X_train, y_train)

#предсказание
y_pred = reg_model.predict(X_test)

# метрики качества
print(f'R2 score: {r2_score(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'Mean absolute percentage error: {mean_absolute_percentage_error(y_test, y_pred)}')
print(f'Explained variance: {explained_variance_score(y_test, y_pred)}')
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(y_pred, y_test)
```

**В целом получается очень даже ничего, модель еще можно было бы докрутить. Но в целом - затея не очень эффективная. Так как эти данные экспрессии никак не нормализированы, а значит как только мы перейдем к другому датасету, собранному в аналогичном эксперименте - модель потеряет свою эффективность.**

### 4. PCA 

```python
# a fuction to scale data and perform PCA with a given number of PC and features
def PC_analysis(df, features, target, n_pc = 10):
    x = df.loc[:, features].values
    scaler = StandardScaler().fit(x)
    x = scaler.transform(x) #rescaled data
    
    pca = PCA(n_components = n_pc)
    pca.fit(x)
    principalComponents = pca.transform(x)
    principalDf = pd.DataFrame(data = principalComponents,
                               columns = [f'PC{i}' for i in range(1, n_pc+1)])
    
    scoresDf = pd.concat([principalDf, df[target]], axis = 1)
    
    loadingsDf = pd.DataFrame(data=pca.components_.T, index=features,
                              columns=[f'PC{i}' for i in range(1, n_pc+1)]) 
    
    return scoresDf, loadingsDf, scaler, pca

# correaltion of features with principal components
def loading_matrix(model, features):
    loadings = model.components_.T * np.sqrt(model.explained_variance_)
    cols = [f'PC{i}' for i in range(1, model.components_.shape[0]+1)]
    loading_matrix = pd.DataFrame(loadings, columns=cols, index=features)
    return loading_matrix
```

**Для PCA не будем проводить удаление колонок с белками, так как это ухудшит разделение точек (проверено самолично). В этом случае мстоит пожертвовать количеством измерений в бользу большего числа фичей**

```python
mice_data_cleaned_hard = mice_data.dropna().reset_index(drop=True)

proteins = list(mice_data_cleaned_hard.iloc[:, 1:-4].columns)
scoresDf, loadingsDf, scaler, pca = PC_analysis(mice_data_cleaned_hard,
                                                features=proteins,
                                                target='class', n_pc = 10)

# add more columns for convenience
scoresDf['Genotype'] =  mice_data_cleaned_hard['Genotype']
scoresDf['Treatment'] =  mice_data_cleaned_hard['Treatment']
```

**Explained variance**

```python
print(f'Explained variance per component: {pca.explained_variance_ratio_*100}')

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

**Loadings**

```python
# plot loadings
def loadings_scatter(Df, pc1 = 1, pc2 = 2, annotate=False, d = 0.01):
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('PC%d' % pc1, fontsize = 10)
    ax.set_ylabel('PC%d' % pc2, fontsize = 10)
    ax.set_title('Loadings', fontsize = 15)
    
    
    features = list(Df.index)
    x = Df.loc[:, 'PC%d' % pc1].values
    y = Df.loc[:, 'PC%d' % pc2].values
    for i in range(len(features)):        
        ax.scatter(x[i], y[i], color='b', s = 30)
        if annotate:
            ax.annotate(features[i], (x[i], y[i]), (x[i] + d, y[i] + d ))

    ax.grid()
```

```python
# можно и аннотировать точки, но тогда ничего не разобрать будет
loadings_scatter(loadingsDf, pc1=1, pc2=2)
```

```python
fig, axes = plt.subplots(2, figsize=(10, 20))

sns.heatmap(loadingsDf, cmap="coolwarm", ax = axes[0])
sns.heatmap(loading_matrix(pca, proteins), cmap="coolwarm", ax = axes[1])
axes[0].set_title('Loadings heatmap')
axes[1].set_title('Correlation of components with features')
plt.show()
```

**Scores**

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 6))
ax = ax.flatten()

sns.scatterplot(data=scoresDf, x='PC1', y='PC2', hue='class', ax=ax[0])
sns.scatterplot(data=scoresDf, x='PC1', y='PC2', hue='Genotype', ax=ax[1])
sns.scatterplot(data=scoresDf, x='PC1', y='PC2', hue='Treatment', ax=ax[2])
plt.show()
```
